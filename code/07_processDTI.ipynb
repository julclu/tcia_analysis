### Find what the name of the series that we have to process is: 

First get rid of anything that doesn't have the word "dti" or "DTI" in it: 

```python
import ast
```

```python
dti_to_process = []
for idx, list_of_dti in enumerate(map_dti_subset['dti']): 
    list_of_dti = ast.literal_eval(list_of_dti)
    if len(list_of_dti) == 1: 
        dti_to_process.append((idx, list_of_dti[0]))
    else: 
        list_of_dti.sort(key = lambda x: x.split('-')[0])
    for elt in list_of_dti: 
        if len(elt.split('-')[0]) == 1 or len(elt.split('-')[0]) == 2 and 'dti' in elt.lower() and 'adc' not in elt.lower() and 'apparent' not in elt.lower(): 
            dti_to_process.append((idx, elt))
            break
#         elif len(elt.split('-')[0]) == 1 and 'dti' in elt.lower() and 'adc' not in elt.lower(): 
#             dti_to_process.append((idx, elt))
#             break
#         elif len(elt.split('-')[0]) == 1 and 'dti' in elt.lower(): 
#             dti_to_process.append((idx, elt))
#             break
#         else: 
#             dti_to_process.append((idx, elt))
len(dti_to_process)            
```

```python
def process_dti_one_exam(root_dir, cohort, patient_id, exam, dti_series_name, path_to_command): 
    path_to_dti = Path(root_dir, cohort, patient_id, exam)
    os.chdir(path_to_dti)
    
    if len([x for x in os.listdir() if 'diffusion' in x]) < 1: 
        command_to_process = path_to_command+'/process_DTI_brain '+dti_series_name+" "+patient_id
        print(command_to_process)
        try: 
            sub.call(command_to_process, shell = True)
        except Exception as error: 
            print(error)
    else: 
        print('Already processed!')
        
```

```python
process_dti_one_exam(root_dir, 
                     cohort, 
                     patient_id, 
                     exam, 
                     dti_series_name, 
                     path_to_command
                    )
```
