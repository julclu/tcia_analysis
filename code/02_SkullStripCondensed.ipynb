## Skullstrip only the niftis that are in the final analysis

For each exam, we want to find the T1, T2, T2FLAIR, and T1C and for the first go, we're just going to skull strip using the BET program from FSL. 

We want to use the following command: 

` bet <input> <output> -S`

```python
import glob
import subprocess as sub 
from subprocess import * 
import os 
from pathlib import Path
import pandas as pd
```

We can use the output from a prior analysis to label our images: 

```python
PATH = Path('/working/lupolab/julia/cnn_analysis/who_subtype')
```

```python
tcia_prenpy = pd.read_csv(Path(PATH, 'data', 'processed_csvs', 'prenpy_tcia.csv'))
```

```python
tcia_prenpy.head()
```

```python
tcia_usable = tcia_prenpy.loc[~tcia_prenpy.label.isnull()]
```

```python
tcia_usable.shape
```

```python
command_premask
```

```python
commands_premask = []
commands_skullstrip = []
for idx, row in tcia_usable.iterrows():
    path_to_exam = row['img_loc']
    
    all_old_vols = glob.glob(str(path_to_exam)+'/*a.nii.gz')
    print(all_old_vols)
    
    for old_vol in all_old_vols: 
        premask_vol = old_vol.split('.n')[0]+'_premask'+'.nii.gz'
        output_vol  = old_vol.split('.n')[0]+'_n'+'.nii.gz'
        command_premask = 'standard_space_roi '+old_vol+' '+premask_vol+' -b'
        command_skullstrip = 'bet '+premask_vol+' '+output_vol+' -f 0.07'
        commands_premask.append(command_premask)
        commands_skullstrip.append(command_skullstrip)
```
